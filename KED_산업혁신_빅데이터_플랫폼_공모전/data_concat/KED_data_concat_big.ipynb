```python
train=pd.read_csv('/content/drive/MyDrive/애쓰는 감자/data/train.csv',dtype={'KEDCD':str,'BZ_PPOS_ITM_CTT':str,'KSIC10_BZC_CD':str,'mid':str,'Big':str})
train.info()
```

```python
# 라벨링
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(trn_shuffle['Big'])
trn_shuffle['Big_idx'] = encoder.transform(trn_shuffle['Big'])
trn_shuffle.head()
```

```python
# 라벨링된 것 매핑
mapping = dict(zip(range(len(encoder.classes_)), encoder.classes_))
mapping
```
